```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.optimize import curve_fit
import scipy.stats as stat
from scipy import asarray as ar,exp
```

```python
#kb=1.38064852*10**(-23)
kb=1.38064852e-5 #nm2.kg/K.s2
T=300

Na=6.02214e23
def gaussian(x,a,b,c):
    func=a*exp(-(x-b)**2/(2*c**2))
    return(func)
```

```python
data=pd.read_csv("angdist2_SNaSNaSNa.xvg", skiprows=13,delim_whitespace=True, names=["Angle","frequency"])
data.head()
```

```python
#from scipy import asarray as ar,exp

x= data['Angle']
y=data['frequency']

n = len(data["Angle"])+1                          #the number of data
x0 = sum((data["Angle"])*(data["frequency"]))                   #note this correction
sigma = sum(y*(x-x0)**2)        #note this correction
print('mean=  ',x0)
print('sigma= ',sigma)
def gaus(x,a,x0,sigma):
    return a*exp(-(x-x0)**2/(2*sigma**2))

popt,pcov = curve_fit(gaus,x,y,p0=[1,x0,sigma])

plt.plot(x,y,'b+:',label='data')
plt.plot(x,gaus(x,*popt),'ro:',label='fit')
plt.legend()
plt.xlabel('Angle of SNaSNaSNa')
plt.ylabel('Frequency')
plt.show()

y_fits=gaussian(x,popt[0],popt[1],popt[2])
std=np.std([x,y_fits])
kb=1.38064852*10**(-23)
T=300
force_const=kb*T*Na/((std**2))
print('force constant= ',force_const) #Angle
```
